```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")



# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
session = Session(engine)
```

```python
# query date and precipitation columns
results = session.query(Measurement.date, Measurement.prcp).all()
```

```python
# 1 Write a query that filters the Measurement table to retrieve the precipitation for the month of June.
from sqlalchemy import extract
session.query(Measurement.date, Measurement.prcp).filter(extract('month',Measurement.date)==6).all() 
```

```python
# retrieve the precipitation for the month of June
df = pd.DataFrame(results, columns=['date','precipitation'])
df.set_index(df['date'], inplace=True)

df = df.sort_index()
df_june = df['2010-06-01':'2016-06-30']
```

```python
df_june.describe()
```

```python
# 2. Write a query that filters the Measurement table to retrieve the precipitation for the month of December. 
from sqlalchemy import extract
session.query(Measurement.date, Measurement.prcp).filter(extract('month',Measurement.date)==12).all()
```

```python
# retrieve the precipitation for the month of December
df = pd.DataFrame(results, columns=['date','precipitation'])
df.set_index(df['date'], inplace=True)

df = df.sort_index()
df_dec = df['2010-12-01':'2016-12-31']
```

```python
df_dec.describe()
```

```python
### Second query related to the frequency of temperature in the station with highest observations as a aditional information
### Related to Jun, 2016
```

```python
# How many stations are available in this dataset?
session.query(func.count(Station.station)).all()
```

```python
session.query(Measurement.station, func.count(Measurement.station)).\
    group_by(Measurement.station).order_by(func.count(Measurement.station).desc()).all()
```

```python
# frequency of temperature in the station with highest observations in June month 

import datetime as dt
from pandas.plotting import table
prev_year = dt.date(2017, 8, 23) - dt.timedelta(days=365)

results = session.query(Measurement.tobs).\
    filter(Measurement.station == 'USC00519281').\
    filter(Measurement.station >= '2016-06-01').\
    filter(Measurement.date <= '2016-06-30').all()
df = pd.DataFrame(results, columns=['tobs'])
df.plot.hist(bins=12)
plt.title('Temperature June vs frequency by USC00519281 Station')
plt.xlabel('Temperature in fahrenheit degree')
plt.tight_layout()
```
